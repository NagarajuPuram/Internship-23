```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
#2 Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release)
```

```python
page=requests.get('https://www.imdb.com/list/ls091520106/')
page
```

```python
soup=BeautifulSoup(page.content)
soup.text
```

```python
#2 IMDB’s Top rated 100 movies
```

```python
movies=[]
for i in soup.find_all('h3',class_="lister-item-header"):
    movies.append(i.text.replace('\n',''))
    
movies
```

```python
ratings=[]
for i in soup.find_all('div',class_="ipl-rating-star small"):
    ratings.append(i.text.replace('\n',""))
    
ratings
```

```python
#Data frame
df = pd.DataFrame()
df['Top 100 movies and their Year of Release'] = movies
df['Ratings'] = ratings

df
```

```python
#4 Product name,Price and discounts from MEESHO
```

```python
page=requests.get('https://meesho.com/bags-ladies/pl/p7vbp')
page
```

```python
!pip install pandas
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
Product_names = []
for i in soup.find_all("p",class_="Text__StyledText-sc-oo0kvp-0 bWSOET NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS"):
    Product_names.append(i.text.replace("\n",""))
Product_names
```

```python
All_Prices = []
for i in soup.find_all("h5",class_="Text__StyledText-sc-oo0kvp-0 hiHdyy"):
    All_Prices.append(i.text.replace("\n",""))
All_Prices
```

```python
All_discounts = []
for i in soup.find_all("p",class_="Text__StyledText-sc-oo0kvp-0 fCJVtz NewProductCard__DiscountTextParagraph-sc-j0e7tu-16 kmYsnm NewProductCard__DiscountTextParagraph-sc-j0e7tu-16 kmYsnm"):
    All_discounts.append(i.text.replace("\n",""))
All_discounts
```

```python
df = pd.DataFrame()
df['Product names'] = Product_names
df['Product prices'] = All_Prices
df['Product discounts'] = All_discounts
df
```

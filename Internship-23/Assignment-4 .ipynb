1. Scrape the details of most viewed videos on YouTube from Wikipedia.

```python
driver = webdriver.Chrome('chromedriver1.exe')
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos' 
driver.get(url)
```

```python
rank = []
name=[]
artist =[]
date=[]
views=[]

r=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[1]')
for i in r:
    rank.append(i.text)
n=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[2]')
for j in n:
    name.append(j.text)
a=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[3]')
for k in a:
    artist.append(k.text)
d=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[5]')
for l in d:
    date.append(l.text)
v=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[4]')
for m in v:
    views.append(m.text)
        
print(len(rank))
print(len(name))
print(len(artist))
print(len(date))
print(len(views))
```

```python
df = pd.DataFrame({'Rank':rank,
                 'Name':name,
                 'Artist':artist,
                 'Date':date,
                 'Views':views})
df
```

5. Scrape the details of trending repositories on Github.com.

```python
url = 'https://github.com/' 
driver.get(url)
```

```python
explore = driver.find_element_by_xpath('/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/summary')
explore.click()
```

```python
trending = driver.find_element_by_xpath('/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul/li[5]/a')
trending.click()
```

```python
title = []
description = []
contributer = []

soup=BeautifulSoup(driver.page_source,'html.parser')
tit = soup.find_all('h1',attrs={'h3 lh-condensed'})
for i in tit:
    try:
        title.append(i.find('span').text.strip('\n'))
    except AttributeError:
        title.append("No details available")
desc = soup.find_all('article',attrs={'Box-row'})
for j in desc:
    try:
        description.append(j.find('p').text.split('\n')[-2])
    except AttributeError:
        description.append("No details available")
cont = soup.find_all('div',attrs={'f6 color-fg-muted mt-2'})
for k in cont:
    try:
        contributer.append(k.find('a').text)
    except AttributeError:
        contributer.append("No details available")
        
print(len(title))
print(len(description))
print(len(contributer))
```

```python
urls = []

links = driver.find_elements_by_xpath('//article[@class="Box-row"]//h1//a')

for url in links:
    urls.append(url.get_attribute('href'))
urls
```

```python
title_of_repository = []
repository_description = []
count_of_contributors = []
language_in_use = []

for i in urls:
    driver.get(i)
    
    try:
        title = driver.find_element_by_xpath('//h1[@class=" d-flex flex-wrap flex-items-center wb-break-word f3 text-normal"]/span[1]')
        title_of_repository.append(title.text.replace('\n',''))
    except NoSuchElementException:
        title_of_repository.append('-')
        
    
    try:
        descrip = driver.find_element_by_xpath('//div[@class="BorderGrid-cell"]//p')
        repository_description.append(descrip.text.replace('\n',''))
    except NoSuchElementException:
        repository_description.append('-')

    
    try:
        count = driver.find_element_by_xpath("//h2[@class='h4 mb-3']/a[contains(text(),'Contributors')]/span")
        count_of_contributors.append(count.text)
        
    except NoSuchElementException:
        count_of_contributors.append('-')
    
    try:
        lang = driver.find_element_by_xpath('//li[@class="d-inline"]//a//span')
        language_in_use = [].append(lang.text.replace('\n',''))
    except NoSuchElementException:
        language_in_use = [].append('-')
```

```python
df5 = pd.DataFrame({})
df5['repository_title'] = title_of_repository
df5['repository_description'] = repository_description
df5['contributors_count']  = count_of_contributors
df5['language_used'] = language_in_use
df5
```

10. Details of Datasets from UCI machine learning repositories.

```python
url = 'https://archive.ics.uci.edu/'
driver.get(url)
```

```python
view = driver.find_element_by_xpath('//span[@class="whitetext"][2]/a[1]')
view.click()
```

```python
name = []
typee = []
task = []
attribute = []
instances = []
attributes = []
year = []

nam = driver.find_elements_by_xpath('//table[@cellpadding = "3"]/tbody/tr/td[2]/table[2]/tbody/tr/td[1]')
typ = driver.find_elements_by_xpath('//table[@cellpadding = "3"]/tbody/tr/td[2]/table[2]/tbody/tr/td[2]')
tas = driver.find_elements_by_xpath('//table[@cellpadding="3"]/tbody/tr/td[2]/table[2]/tbody/tr/td[3]')
attri = driver.find_elements_by_xpath('//table[@cellpadding="3"]/tbody/tr/td[2]/table[2]/tbody/tr/td[4]')
inst = driver.find_elements_by_xpath('//table[@cellpadding="3"]/tbody/tr/td[2]/table[2]/tbody/tr/td[5]')
attri = driver.find_elements_by_xpath('//table[@cellpadding="3"]/tbody/tr/td[2]/table[2]/tbody/tr/td[6]')
yea = driver.find_elements_by_xpath('//table[@cellpadding="3"]/tbody/tr/td[2]/table[2]/tbody/tr/td[7]')

for i in nam[1:]:
    name.append(i.text)
for j in typ[1:]:
    typee.append(j.text)
for k in tas[1:]:
    task.append(k.text)
for l in attri[1:]:
    attribute.append(l.text)
for m in inst[1:]:
    instances.append(m.text)
for n in attri[1:]:
    attributes.append(n.text)
for o in yea[1:]:
    year.append(o.text)
    
print(len(name))
print(len(typee))
print(len(task))
print(len(attribute))
print(len(instances))
print(len(attributes))
print(len(year))
```

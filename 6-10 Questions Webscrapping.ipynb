9A. Dineout.co.in :
i) Restaurant name
ii) Cuisine
iii) Location
iv) Ratings
v) Image URL

```python
page=requests.get("https://www.dineout.co.in/delhi-restaurants/buffet-special")
page
```

```python
page_cnt=bs(page.content)
page_cnt
```

```python
hotel_name=[]
for i in page_cnt.find_all("div",class_=("restnt-info cursor")):
    hotel_name.append(i.text)
hotel_name
```

```python
loc=[]
for i in page_cnt.find_all("div",class_=("restnt-loc ellipsis")):
    loc.append(i.text)
loc
```

```python
price_cuisine=[]
for i in page_cnt.find_all("span", class_="double-line-ellipsis"):
    price_cuisine.append(i.text)
price_cuisine
```

```python
rating=[]
for i in page_cnt.find_all("div", class_="restnt-rating rating-4"):
    rating.append(i.text)
rating
```

```python
image_url=[]
for i in page_cnt.find_all("img",class_="no-img"):
    image_url.append(i['data-src'])
image_url
```

```python
import pandas as pd
df=pd.DataFrame({'Rest_name':hotel_name,"location":loc,"price & cuisine":price_cuisine,"Rating":rating,"img_url":image_url,})
df
```

```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

```python
page=requests.get("https://www.bewakoof.com/women-tshirts?ga_q=tshirts")
page
```

```python
soup=bs(page.content)
soup
```

```python
Product_name = []  # empty list to store
 
for i in soup.find_all("div",class_="productCardDetail"):
    Product_name.append(i.text.split('|'))

Product_name
```

```python
Price = []

for i in soup.find_all("span",class_="discountedPriceText"):
    Price.append(i.text)
    

Price
```

```python
images = []

for i in soup.find_all("img",class_="productImgTag"):
    images.append(i['src'])
images
```

```python
df=pd.DataFrame({'Product Name':Product_name,'Price':Price,'Images_url':images})
df
```
